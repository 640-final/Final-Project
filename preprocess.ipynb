```python
import json
import pandas as pd
```

```python
f1 = open('Twitter_User_Handles_labeled_tweets.json')
data_tweets = json.load(f1).items()
f2 = open('User_demo_profiles.json')
data_profile = json.load(f2)
user_label = pd.read_csv('labeled_users.csv')
```

```python
user_label.rename(columns={"user_id": "id"}, inplace = True)
user_label["id"] = user_label["id"].astype(int)
user_label.head()
```

```python
data_profile = pd.DataFrame(data_profile)
data_profile = data_profile[["id", "name", "screen_name", "description", "img_path"]]
# data_profile = data_profile.set_index(["id"])
data_profile.head()
```

```python
data_tweets = pd.DataFrame.from_dict(data_tweets)
data_tweets.rename(columns={0: "id", 1: "tweets"}, inplace = True)
data_tweets["id"] = data_tweets["id"].astype(int)
# data_tweets = data_tweets.set_index(["id"])
data_tweets[:5]
```

```python
alldata = pd.merge(data_tweets, data_profile, on='id')
# alldata = pd.concat([data_tweets, data_profile], axis=1)
alldata = pd.merge(alldata, user_label, on='id')
alldata[:5]
```

```python
alldata.to_csv("data_race_pred.csv")
```
